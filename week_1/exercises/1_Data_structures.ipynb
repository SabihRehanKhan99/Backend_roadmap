### Duplicate Detector

```python
def has_duplicate(items):

    seen = set()

    for item in items:

        if item in seen:
            return True

        seen.add(item)

    return False



print(has_duplicate([1, 2, 3, 4, 5]))  # Output: False
print(has_duplicate([1, 2, 3, 4, 5, 2]))  # Output: True
```

### Frequency Counter

```python
def frequency_counter(items):

    freq = {}

    for item in items:

        if item in freq:
            freq[item] += 1
        else:
            freq[item] = 1

    return freq


data = ["apple","banana","apple","orange","banana","apple"]

print(frequency_counter(data))
```

### LRU Cache

```python
from collections import OrderedDict

class LRUCache:

    def __init__(self, capacity):
        self.capacity = capacity
        self.cache = OrderedDict()

    def get(self, key):

        if key not in self.cache:
            return -1

        self.cache.move_to_end(key)

        return self.cache[key]

    def put(self, key, value):

        if key in self.cache:
            self.cache.move_to_end(key)

        self.cache[key] = value

        if len(self.cache) > self.capacity:
            self.cache.popitem(last=False)
```

```python
cache = LRUCache(3)

cache.put("A",1)
cache.put("B",2)
cache.put("C",3)

cache.get("A")

cache.put("D",4)

print(cache.cache)
```

### Nested Configuration Parser

```python
def merge_configs(base, override):

    for key, value in override.items():

        if key in base and isinstance(base[key], dict) and isinstance(value, dict):
            merge_configs(base[key], value)

        else:
            base[key] = value

    return base


default_config = {
    "database": {
        "host": "localhost",
        "port": 5432
    },
    "cache": {
        "enabled": True,
        "ttl": 300
    }
}

file_config = {
    "database": {
        "host": "db.production.com"
    }
}

env_config = {
    "cache": {
        "ttl": 600
    }
}

config = merge_configs(default_config.copy(), file_config)
config = merge_configs(config, env_config)

print(config)
```

### Task Priority Queue

```python
from queue import PriorityQueue


class Task:

    def __init__(self, priority, description):
        self.priority = priority
        self.description = description

    def __lt__(self, other):
        return self.priority < other.priority


class TaskQueue:

    def __init__(self):
        self.queue = PriorityQueue()

    def add_task(self, priority, description):

        task = Task(priority, description)

        self.queue.put(task)

        print(f"Task added: {description} (priority {priority})")

    def process_task(self):

        if self.queue.empty():
            print("No tasks left")
            return

        task = self.queue.get()

        print(f"Processing task: {task.description}")
```

```python
task_queue = TaskQueue()

task_queue.add_task(3, "send email")
task_queue.add_task(1, "process payment")
task_queue.add_task(2, "generate report")

task_queue.process_task()
task_queue.process_task()
task_queue.process_task()
```

### User Entity with Validation

```python
from dataclasses import dataclass, field
from datetime import datetime
import re

@dataclass
class User:

    id: int
    name: str
    email: str
    age: int
    is_active: bool = True
    created_at: datetime = field(default_factory=datetime.now)

    def __post_init__(self):

        if not self.name.strip():
            raise ValueError("Name cannot be empty")

        if self.age < 0:
            raise ValueError("Age cannot be negative")

        if not re.match(r"[^@]+@[^@]+\.[^@]+", self.email):
            raise ValueError("Invalid email address")
        


user = User(
    id=1,
    name="Ali",
    email="ali@gmail.com",
    age=25
)

print(user)
```
